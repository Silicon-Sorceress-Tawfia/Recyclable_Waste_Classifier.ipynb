```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set up the data generators
train_datagen = ImageDataGenerator(rescale=1./255)  # Rescale images to 0-1 range
validation_datagen = ImageDataGenerator(rescale=1./255)

# Define the training generator
train_generator = train_datagen.flow_from_directory(
    'data/train',  # Path to the training data
    target_size=(150, 150),  # Resize images to 150x150 pixels
    batch_size=32,  # Number of images to process in a batch
    class_mode='categorical')  # Use categorical labels

# Define the validation generator
validation_generator = validation_datagen.flow_from_directory(
    'data/validation',  # Path to the validation data
    target_size=(150, 150),  # Resize images to 150x150 pixels
    batch_size=32,  # Number of images to process in a batch
    class_mode='categorical')  # Use categorical labels
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Define a simple CNN model
def create_model():
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
        MaxPooling2D(2, 2),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(5, activation='softmax')  # Adjust the number of classes
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

model = create_model()
model.summary()  # Print the model summary

# Train the model
model.fit(train_generator, epochs=10, validation_data=validation_generator)
```

```python
model.save('waste_classifier_model.h5')
```
